This Project visualizes the Gender gap in college degrees between the years 1970 to 2012.

In this project, we'll be visualizing the gender gap across college degrees. We will generate line charts to compare across all degree categories. In the last step of this project, we'll explore how to export the final diagram we create as an image file.

Because there are seventeen degrees that we need to generate line charts for, we'll use a subplot grid layout of 6 rows by 3 columns. We can then group the degrees into STEM, liberal arts, and other, in the following way:

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 
             'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                 'Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 
              'Education', 'Agriculture','Business', 'Architecture']
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('../Dataset/percent-bachelors-degrees-women-usa.csv')
women_degrees.head(3)
```

## Generating line charts for all three subject categories of 

```python
# set the size of the figure container
fig = plt.figure(figsize = (18, 25))

# set color of the lines for males and females
female_color = (0/255, 107/255, 164/255) 
male_color = (255/255, 128/255, 14/255)

# charts for the STEM subjects
for i in range(0, 18, 3): # this returns the numbers (0, 3, 6, 9, 12, 15) and skips two columns before moving to the next line
    cat_index = int(i / 3) # this returns the numbers (0, 1, 2, 3, 4, 5) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    # remove the axis ticks from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False)
    
    # set title of charts
    ax.set_title(stem_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2004, 83, 'Women')
        ax.text(2008, 15, 'Men')
    elif cat_index == 5:
        ax.text(2008, 90, 'Men')
        ax.text(2004, 10, 'Women')

# Chatrs for the Library and Arts subjects
for i in range(1, 16, 3): # this returns the numbers (1, 4, 7, 10, 13) and skips two columns
    cat_index = int((i - 1) / 3) # this returns the numbers (0, 1, 2, 3, 4) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # remove the axis ticks from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False)
    
    # set title of charts
    ax.set_title(lib_arts_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2006, 77, 'Women')
        ax.text(2008, 20, 'Men')
    elif cat_index == 4:
        ax.text(1973, 65, 'Men')
        ax.text(1973, 30, 'Women')
        
# Charts for Other Categories
for i in range(2, 18, 3): # this returns the numbers (2, 5, 8, 11, 14, 17) and skips two columns
    cat_index = int((i-2) / 3) # this returns the numbers (0, 1, 2, 3, 4, 5) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # remove the axis ticks from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False)
    
    # set title of charts
    ax.set_title(other_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2004, 93, 'Women')
        ax.text(2008, 8, 'Men')
    elif cat_index == 5:
        ax.text(2008, 62, 'Men')
        ax.text(2004, 36, 'Women')
plt.show()
```

## Removing x-axis Labels for all Columns except for the Bottom-Most Chart

With seventeen line charts in one diagram, the non-data elements quickly clutter the field of view. The most immediate issue that sticks out is the titles of some line charts overlapping with the x-axis labels for the line chart above it. If we remove the titles for each line chart, the viewer won't know what degree each line chart refers to. Let's instead remove the x-axis labels for every line chart in a column except for the bottom most one.

```python
# set the size of the figure container
fig = plt.figure(figsize = (18, 25))

# set color of the lines for males and females
female_color = (0/255, 107/255, 164/255) 
male_color = (255/255, 128/255, 14/255)

# charts for the STEM subjects
for i in range(0, 18, 3): # this returns the numbers (0, 3, 6, 9, 12, 15) and skips two columns before moving to the next line
    cat_index = int(i / 3) # this returns the numbers (0, 1, 2, 3, 4, 5) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    # remove the axis ticks and x-axis bottom labels from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False, labelbottom = False)
    
    # set title of charts
    ax.set_title(stem_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2004, 83, 'Women')
        ax.text(2008, 15, 'Men')
    elif cat_index == 5:
        ax.text(2008, 90, 'Men')
        ax.text(2004, 10, 'Women')
        ax.tick_params(labelbottom = True) # Adding the bottom label for the last chart in a column

# Chatrs for the Library and Arts subjects
for i in range(1, 16, 3): # this returns the numbers (1, 4, 7, 10, 13) and skips two columns
    cat_index = int((i - 1) / 3) # this returns the numbers (0, 1, 2, 3, 4) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # remove the axis ticks and x-axis bottom labels from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False, labelbottom = False)
    
    # set title of charts
    ax.set_title(lib_arts_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2006, 77, 'Women')
        ax.text(2008, 20, 'Men')
    elif cat_index == 4:
        ax.text(1973, 65, 'Men')
        ax.text(1973, 30, 'Women')
        ax.tick_params(labelbottom = True) # Adding the bottom label for the last chart in a column
        
# Charts for Other Categories
for i in range(2, 18, 3): # this returns the numbers (2, 5, 8, 11, 14, 17) and skips two columns
    cat_index = int((i-2) / 3) # this returns the numbers (0, 1, 2, 3, 4, 5) and uses it to index
    ax = fig.add_subplot(6, 3, i + 1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c = female_color, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[cat_index]], c = male_color, linewidth = 3, label = 'Men')
    
    # set axis limits
    ax.set_ylim(0,100)
    ax.set_xlim(1970, 2012)
    
    # remove axis spines from charts
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # remove the axis ticks and x-axis bottom labels from charts
    ax.tick_params(bottom = False, top = False, right = False, left = False, labelbottom = False)
    
    
    # set title of charts
    ax.set_title(other_cats[cat_index])
    
    # set annotation for top and bottom charts
    if cat_index == 0:
        ax.text(2004, 93, 'Women')
        ax.text(2008, 8, 'Men')
    elif cat_index == 5:
        ax.text(2008, 62, 'Men')
        ax.text(2004, 36, 'Women')
        ax.tick_params(labelbottom = True) # Adding the bottom label for the last chart in a column
plt.show()
```

Removing the x-axis labels for all but the bottommost plots solved the issue we noticed with the overlapping text. In addition, the plots are cleaner and more readable. The trade-off we made is that it's now more difficult for the viewer to discern approximately which years some interesting changes in trends may have happened. This is acceptable because we're primarily interested in enabling the viewer to quickly get a high level understanding of which degrees are prone to gender imbalance and how that has changed over time.

In the vein of reducing cluttering, let's also simplify the y-axis labels. Currently, all seventeen plots have six y-axis labels and even though they are consistent across the plots, they still add to the visual clutter. By keeping just the starting and ending labels (0 and 100), we can keep some of the benefits of having the y-axis labels.

While removing most of the y-axis labels definitely reduced clutter, it also made it hard to understand which degrees have close to 50-50 gender breakdown. While keeping all of the y-axis labels would have made it easier, we can actually do one better and use a horizontal line across all of the line charts where the y-axis label 50 would have been.

We can generate a horizontal line across an entire subplot using the [Axes.axhline()](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.axhline.html#matplotlib.axes.Axes.axhline) method. The only required parameter is the y-axis location for the start of the line:

We'll also use the next color in the [Color Blind 10 palette](http://tableaufriction.blogspot.com/2012/11/finally-you-can-use-tableau-data-colors.html) for this horizontal line, which has an RGB value of (171, 171, 171). Because we don't want this line to clutter the viewing experience, We'll increase the transparency of the line. We can set the color using the **c** parameter and the transparency using the **alpha** parameter. The value passed in to the alpha parameter must range between 0 and 1

## Add a Horizontal line That Shows a 50% Breakdown of Genders in Each Subject
