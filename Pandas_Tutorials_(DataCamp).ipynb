```python
# check
list_values
```

The above output shows that the `list_values` is a tuple of two lists `country` and `total`



```python
# # Zip the lists and tuple together into one zip file (key,value): zipped
zipped = zip(list_keys, list_values)

## Inspect the list using print()
zipped
```

```python
# Build a dictionary with the zipped list: data
data = dict(zipped)
data
```

```python
# Build and inspect a DataFrame from the dictionary: df
df = pd.DataFrame(data)
df.head(3)
```

### pandas scatter plots


```python
uploaded7 = files.upload()
```

```python
# import the dataset for the task
auto_mobile = pd.read_csv('auto-mpg.csv')

# check
auto_mobile.head()
```

```python
# get a numpy array of car weight called weight
weight = np.array(auto_mobile['weight'])

# get the mean of the weights
mean_weight = weight.mean()

# normalize the array, Divide an array by its mean
sizes = weight/mean_weight

# check
sizes
```

```python
# Generate a scatter plot
auto_mobile.plot(kind='scatter', x='hp', y='mpg', s=sizes)

# Add the title
plt.title('Fuel efficiency vs Horse-power')

# Add the x-axis label
plt.xlabel('Horse-power')

# Add the y-axis label
plt.ylabel('Fuel efficiency (mpg)')

# Display the plot
plt.show()
```

### pandas box plots


```python
# Make a list of the column names to be plotted: cols
cols = ['weight', 'mpg']

# Generate the box plots
auto_mobile[cols].plot(kind='box', subplots=True, figsize = (8,4))

# Display the plot
plt.show()
```

# Filtering and counting


```python
# Lets import the dataset required for this task
uploaded12 = files.upload()
```

```python
# get the data text
data_text3 = 'automobiles.csv'

# read the dataset as a csv file
cars = pd.read_csv(data_text3)

# lets inspect the dataset required for this task
cars.head()
```

```python
# count cars manufactured in  asia
asia = cars[cars['origin'] == 'Asia']

asia.count()
```

# Separate and summarize


```python
# Compute the global mean and global standard deviation of all cars: global_mean, global_std
global_mean = cars.mean()
global_std = cars.std()

# Filter the US population from the origin column: us
us = cars[cars['origin']=='US']

# Compute the US mean and US standard deviation: us_mean, us_std
us_mean = us.mean()
us_std = us.std()

# Print the differences
print(us_mean - global_mean, '\n')
print(us_std - global_std)
```
